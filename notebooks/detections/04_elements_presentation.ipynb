```python

CONFIG = SoccerPitchConfiguration()
annotated_frame = draw_pitch(CONFIG)

trajectory_color = sv.Color.RED
ball_color = sv.Color.BLUE

for col in df.columns:
    if col == 'frame_index':
        continue
    # Drop NaN values and replace semicolons with commas before evaluating
    col_xy = (
        df[col]
        .dropna()
        .str.replace(';', ',')
        .apply(ast.literal_eval)
        .to_list()
    )
    pitch_col_xy = np.array(col_xy)
    # Use blue for the ball, red for others
    color = ball_color if col == 'ball' else trajectory_color
    annotated_frame = draw_paths_on_pitch(
        config=CONFIG,
        paths=[pitch_col_xy],
        color=color,
        thickness=4,
        pitch=annotated_frame
    )

sv.plot_image(annotated_frame)
```
