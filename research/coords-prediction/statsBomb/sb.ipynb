```python
import matplotlib.pyplot as plt
from mplsoccer import Pitch
import pandas as pd

# Load the data
data = pd.read_csv('../../data/raw/subset_1.csv')
```

```python
# Plot ball coordinates on a 120x70 pitch

pitch = Pitch(pitch_length=120, pitch_width=70)
fig, ax = pitch.draw()

# Replace 'ball_x' and 'ball_y' with your actual column names
ax.scatter(data['ball_x'], data['ball_y'], color='red', s=20, label='Ball position')

plt.legend()
plt.title('Ball Coordinates on 120x70 Pitch')
plt.show()
```
